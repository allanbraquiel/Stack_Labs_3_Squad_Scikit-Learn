### Calculando a Taxa de Retorno Simples

RS = (preço final - preço inicial) / preço inicial) * 100

Exemplo: 
preço inicial(compra): 14,99
preço final (venda): 16,07

RS = ((16,07 - 14,99) / 14,99) * 100 = 7,20

Resultado: 7,2 %

```python
df_concat
```

```python
# Primeiro registro
df_concat["Bitcoin"][0]
```

```python
# Ultimo registro
df_concat["Bitcoin"][len(df_concat) -1]
```

```python
# Retorno Simples(RS) = (preço final - preço inicial) / preço inicial) * 100
((df_concat["Bitcoin"][len(df_concat) -1] - df_concat["Bitcoin"][0]) / df_concat["Bitcoin"][0]) * 100
```

```python
# calculando de forma diferente mais simples
# (preço final / preço inicial -1) * 100
(df_concat["Bitcoin"][len(df_concat) -1] / df_concat["Bitcoin"][0] - 1) * 100
```

### Taxa de retorno para cada criptomoeda

```python
# Criando uma coluna com a taxa de retorno diária para a moeda Bitcoin
df_concat["Bitcoin RS"] = (df_concat["Bitcoin"] / df_concat["Bitcoin"].shift(1)) -1
df_concat.head()
```

```python
# avaliando a taxa de retorno diária
df_concat["Bitcoin RS"].plot()
```

```python
# media da taxa de retorno
df_concat["Bitcoin RS"].mean()
```

```python
# Taxa de retorno anual dias uteis
(df_concat["Bitcoin RS"].mean() * 246) * 100
```

```python
# Calculando a taxa de retorno para cada criptomoeda
df_concat["Doge RS"] = (df_concat["Doge"] / df_concat["Doge"].shift(1)) -1
df_concat["Ethereum RS"] = (df_concat["Ethereum"] / df_concat["Ethereum"].shift(1)) -1
df_concat["Litcoin RS"] = (df_concat["Litcoin"] / df_concat["Litcoin"].shift(1)) -1
df_concat["Solana RS"] = (df_concat["Solana"] / df_concat["Solana"].shift(1)) -1

df_concat.head()
```

### Retorno Logaritmo

O Retorno logaritmo serve para calcular o retorno comparando com a propria moeda

RL = log(preço final / preço inicial) * 100

```python
# taxa de retorno simples
(df_concat["Bitcoin"][len(df_concat) -1] / df_concat["Bitcoin"][0] - 1) * 100
```

```python
# taxa de retorno logaritmo
np.log(df_concat["Bitcoin"][len(df_concat) -1] / df_concat["Bitcoin"][0]) * 100
```

```python
# Calculando a taxa de retorno logaritimo para cada criptomoeda e criando uma nova coluna
df_concat["Bitcoin RL"] = np.log(df_concat["Bitcoin"] / df_concat["Bitcoin"].shift(1))
df_concat["Doge RL"] = np.log(df_concat["Doge"] / df_concat["Doge"].shift(1))
df_concat["Ethereum RL"] = np.log(df_concat["Ethereum"] / df_concat["Ethereum"].shift(1))
df_concat["Litcoin RL"] = np.log(df_concat["Litcoin"] / df_concat["Litcoin"].shift(1))
df_concat["Solana RL"] = np.log(df_concat["Solana"] / df_concat["Solana"].shift(1))

df_concat.head()
```
